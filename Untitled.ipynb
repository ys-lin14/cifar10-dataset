```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

from sklearn.model_selection import train_test_split
```

```python
plt.style.use("ggplot")
```

```python
def unpickle(file):
    """Load CIFAR-10 data from a pickle file into a dictionary. 
    Code from https://www.cs.toronto.edu/~kriz/cifar.html
    
    Args: 
        file (str): path to data
    
    Returns:
        data (dict): CIFAR-10 data
    """
    
    with open(file, "rb") as f:
        data = pickle.load(f, encoding="bytes")
    return data

def array_to_image(array):
    """Convert a 1d array back into the array representation 
    of an image that can be shown using plt.imshow(). Based 
    on https://stackoverflow.com/questions/58778364
    
    Args: 
        array (ndarray): 
            1D numpy array of the original image
        
    Returns:
        image (ndarray):
            numpy array representing a 32x32 pixel image with 3 channels 
    """
    
    image = array.reshape(3, 32, 32)
    image = image.transpose(1, 2, 0)
    return image

def display_image(image, axis):
    """Display an image on a given axis without xtick or ytick labels
    
    Args:
        image (ndarray):
            array representing a 32x32 pixel image with 3 channels
        axis (matplotlib subplot axis): 
            axis for which the image is to be plotted on
            
    Returns:
        None
    """
    
    axis.imshow(image)
    axis.set_xticks([])
    axis.set_yticks([])
```

```python
data = unpickle("data/data_batch_1")
print("Keys:", data.keys())
print("Labels:", set(data[b'labels']))
print("(Rows, Columns):", data[b'data'].shape)
```

```python
classes = [
    "airplane", "automobile", "bird", "cat", "deer", 
    "dog", "frog", "horse", "ship", "truck"
]
```

```python
X, y = data[b'data'], np.array(data[b'labels'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
example_indices = np.unique(y_train, return_index=True)[1]
number_examples = len(example_indices)

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15, 3))
axes = axes.flatten()

for i, j in zip(example_indices, range(number_examples)):
    image = array_to_image(X_train[i])
    axes[j].set_title(classes[y_train[i]])
    display_image(image, axes[j])

plt.subplots_adjust(hspace=1)
plt.show()
```
